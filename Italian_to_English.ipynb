```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-it-en")

def preprocess_function(batch):
    inputs = [example["it"] for example in batch["translation"]]
    targets = [example["en"] for example in batch["translation"]]
    model_inputs = tokenizer(inputs, max_length=128, truncation=True)
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True)
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# 1. Split the dataset into 90% train, 10% validation
split_dataset = train_data_set.train_test_split(test_size=0.2, seed=42)
train_data = split_dataset['train']
val_data = split_dataset['test']

# 2. Tokenize both splits
tokenized_train = train_data.map(preprocess_function, batched=True)
tokenized_val = val_data.map(preprocess_function, batched=True)

#train de full model
```

Select a Pre-trained model it-en

```python
from transformers import AutoModelForSeq2SeqLM

# Load the pre-trained MarianMT model for EN→IT translation
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-it-en")
```

```python
from transformers import TrainingArguments
from transformers import DataCollatorForSeq2Seq

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=8,
    num_train_epochs=3,
    learning_rate=5e-5,
    logging_steps=100,
    save_steps=500,
    save_total_limit=2,
    fp16=True,  # Set to False if no GPU
)

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
import os
os.environ["WANDB_DISABLED"] = "false"
```

```python
# Make a tiny train dataset
from transformers import Trainer

#small_train = tokenized_train.select(range(100))   # swap tokenized_train with small_train

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    tokenizer=tokenizer,
    data_collator=data_collator,
)
trainer.train()

```

Evaluate Model

```python
results = trainer.evaluate(eval_dataset=tokenized_val)
print(results)
```
