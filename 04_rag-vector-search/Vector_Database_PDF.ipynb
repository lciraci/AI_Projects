# Ingest PDF

```python
def load_pdf(path):
    chunks = []
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text:
                chunks.append(text)
    return chunks
```

# Embed chunks into FAISS

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pdfplumber
import faiss

model = SentenceTransformer("all-MiniLM-L6-v2")
chunks = load_pdf("CV_CIRACI_LUCIO.pdf")
embeddings = model.encode(chunks, convert_to_numpy=True)

index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings.astype("float32"))
```

# USER QUERY → EMBED → SEARCH FAISS

```python
question = "What should I write in my cover letter?"
q_embed = model.encode([question], convert_to_numpy=True).astype("float32")
D, I = index.search(q_embed, k=1)

context = chunks[I[0][0]]
```

# PASS TO LLM FOR ANSWER

```python
from transformers import pipeline
rag_pipeline = pipeline("text-generation", model="tiiuae/falcon-7b-instruct", device_map="auto")

prompt = f"Context: {context}\n\nQuestion: {question}\nAnswer:"
response = rag_pipeline(prompt, max_new_tokens=100, pad_token_id=rag_pipeline.tokenizer.eos_token_id)[0]['generated_text']

print("\n\U0001F50E Answer:\n", response.encode("utf-8", "ignore").decode("utf-8"))
```
